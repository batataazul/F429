```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
root_path = "~/Documentos/Tarefas/Universidade/F429/Experimento5/CSV/"

files = ["csv1.csv","csv2.csv"]
save = True
```

```python
def ler(i):
    data = pd.read_csv(root_path + i)
    ponto = np.array(data["ponto"])
    tempo = np.array(data["tempo"])
    leitura = np.array(data["leitura"])
    marcacao = np.array(data["marcacao"])
    print("arquivo: " + i)
    return ponto,tempo,leitura,marcacao
```

```python
def plotar(tempo,leitura):
    plt.plot(tempo, leitura,label = "Experimental" )
    plt.xlabel('tempo(s)',fontsize = 20)
    plt.ylabel('intensidade',fontsize = 20)
    plt.legend()
    plt.grid(True)
    if save:
        plt.savefig(root_path + i[0:-4] +" pontos.png",dpi = 1000)  
    plt.show()
    
```

```python
for i in files:
    ponto,tempo,leitura,marcacao = ler(i)
    plotar(tempo,leitura)
```
