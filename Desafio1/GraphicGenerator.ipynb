```python
def pega_dados2(root_path,i):
  data = pd.read_csv(root_path + i)
  time = data["Tempo (s)"].to_numpy(dtype="float64")
  v1 = data["V1 (V)"].to_numpy(dtype="float64")
  v2 = data["V2 (V)"].to_numpy(dtype="float64")
  dif = data["V1 - V2 (V)"].to_numpy(dtype="float64")
  time2 = []
  v12 = []
  v22 = []
  dif2 = []
  for j in np.nditer(time):
    time2.append(j)
  for j in np.nditer(v1):
    v12.append(j)
  for j in np.nditer(v2):
    v22.append(j)
  for j in np.nditer(dif):
    dif2.append(j)
  
  print("arquivo: " + i)
  return time2,v12,v22,dif2
```

```python
def plota_tensao(time,v1,v2,save,i):
  line1, =plt.plot(time[0:-1], v1[0:-1],label = "Tensão de Entrada")
  plt.xlabel('tempo (s)',fontsize = 20)
  plt.ylabel('Tensão (V)',fontsize = 20)
  line2, = plt.plot(time[0:-1], v2[0:-1],label = "tensão de Saída")
  plt.grid(True)
  line3 = plt.vlines(0.0004, -2, 5, color = "purple", label="Resposta transiente")
  line4 = plt.vlines(0.0007, -2, 5, color = "black", label="Resposta Permanente") 
  handle = [line1,line2,line3,line4]
  plt.legend(handles = handle)
  nome = i.split("-")
  plt.title(nome[0],fontsize = 20)
  if save:
    plt.savefig(root_path + i[0:-4] + " tensao.png",dpi = 1000)  
  
  plt.show()
```
