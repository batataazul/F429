```python
import json
import matplotlib.pyplot as plt
```

```python
try:
    arq = open("espectro.json","r")
    print("deu bom")
    j = arq.read()
except:
    print("deu ruim")
arq.close()
```

```python
dic = json.loads(j)
```

```python
for i in dic["data"]:
    plt.plot(converte(i["x"]),converte(i["y"]),label = i["name"],color = i["line"]["color"])
    plt.xlabel('comprimento de onda(nm)',fontsize = 20)
    plt.ylabel('intensidade(arbitrária)',fontsize = 20)
    plt.legend()
plt.savefig("espectro.png", dpi = 1200)    
plt.show()
```

```python
def converte(vetor):
    novo = []
    for i in vetor:
        try:
            novo.append(float(i))
        except:
            print("deu ruim")
            return None
    return novo
```
