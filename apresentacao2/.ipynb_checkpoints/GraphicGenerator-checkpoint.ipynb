```python
root_path = "/content/gdrive/Shared drives/F429/Exp 1/Parte 5/"

'''
files = ["varredura_0,22uF_dados.csv", 
        "varredura_2.2uF.csv",
        "varredura_22uF_dados.csv"]
        
'''


files = ["5 100_dados.csv", "5 10k_dados.csv", "5 5000_dados.csv"]
save = True
```

```python
for i in files:
  
  data = pd.read_csv(root_path + i)

  freq = data["frequencia (Hz)"].to_numpy(dtype="float64")
  v_out = data["Vpp2 (V)"].to_numpy(dtype="float64")
  v_in = data["Vpp1 (V)"].to_numpy(dtype="float64")
  fase = data["fase (Ch2-Ch1) (graus)"].to_numpy(dtype="float64")
  Tdb = data["T_dB"].to_numpy(dtype="float64")
  
  print("arquivo: " + i)
  
  
  #Vout/Vin
  plt.plot(freq[0:-1], v_out[0:-1]/v_in[0:-1], "bo")
  plt.xlabel('Frequência (Hz)')
  plt.ylabel('Tensão de saída / Tensão de entrada')
  
  if save:
    plt.savefig(root_path + i[0:-4] + " v_out v_in.png")
  
  plt.show()
  
  
  #Fase
  plt.plot(freq, fase, "bo")
  plt.xlabel('Frequência (Hz)')
  plt.ylabel('Fase (graus)')
  
  if save:
    plt.savefig(root_path + i[0:-4] + " Fase.png")  
  
  plt.show()
  
  
  #freq
  plt.plot(freq, v_out, "bo")
  plt.xlabel('Frequência (Hz)')
  plt.ylabel('Tensão pico a pico de saída (V)')
  
  if save:
    plt.savefig(root_path + i[0:-4] + " V pico a pico.png")    
  
  plt.show()

  
  #Transmitancia
  plt.plot(freq[0:8], Tdb[0:8], "bo")
  plt.xlabel('Frequência (Hz)')
  plt.ylabel('Transmitância (dB)')
  
  if save:
    plt.savefig(root_path + i[0:-4] + " Transmitancia.png")    
  
  plt.show()
```
