```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,40 * np.pi, 256, endpoint=True)
C, S = np.cos(X), np.cos(0.95*X)
```

```python
som = C[0:] + S[0:] 
```

```python
plt.plot(X,som,label = "batimento")
plt.xlabel('x',fontsize = 20)
plt.ylabel('y',fontsize = 20)
plt.legend()
plt.savefig("batimento.png", dpi = 1200)    
plt.show()
```

```python
plt.plot(X,C,label = "cos(x)",color = "#55f6b7")
plt.xlabel('x',fontsize = 20)
plt.ylabel('y',fontsize = 20)
plt.legend()
plt.savefig("cosx.png", dpi = 1200)    
plt.show()
```

```python
plt.plot(X,S,label = "cos(0.95x)",color = "#db7823")
plt.xlabel('x',fontsize = 20)
plt.ylabel('y',fontsize = 20)
plt.legend()
plt.savefig("cos095x.png", dpi = 1200)    
plt.show()
```
